```python
import jax, jax.numpy as jnp

from untangle.algorithm import cmtf_ssd
from untangle.scaler import JacobianScaler
from untangle.utils import collect_information, function_error

key = jax.random.key(0)
```

```python
def f(x):
    return jnp.array([
        jnp.sin(4*jnp.pi*x[0]) + x[1],
        jnp.cos(3*jnp.pi*x[1]) * x[0]
    ])

m = 2
n = 2
N = 100
rank = 2
```

```python
X, Y, J = collect_information(f, N, m, key=key)

scaler = JacobianScaler(J, Y)
J_scaled, Y_scaled = scaler.scale()
```

```python
learned_scaled = cmtf_ssd(X, Y_scaled, J_scaled, rank, lam=1.0, key=key, iterations=20)
learned = scaler.unscale(learned_scaled)
```
